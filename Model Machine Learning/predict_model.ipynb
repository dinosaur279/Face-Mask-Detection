```python
import matplotlib.pyplot as plt
import dlib
from keras.preprocessing import image
import numpy as np
from os import listdir
from os.path import isfile, join
from keras.models import load_model
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
import cv2
```

```python
# dimensions of our images
img_width, img_height = 150, 150

# load the model we saved
model = load_model('mask_detector_xin.model')
model.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])

mypath = "predict/"
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
print(onlyfiles)
# predicting images
mask = 0 
not_mask  = 0
for file in onlyfiles:
  detector = dlib.get_frontal_face_detector()
  frame = cv2.imread(mypath+file)
  RGB_im = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
  gray =cv2.cvtColor(RGB_im,cv2.COLOR_BGR2GRAY)
  faces = detector(gray)
  fit =1
  for counter,face in enumerate(faces):
    x1, y1 = face.left(), face.top()
    x2, y2 = face.right(), face.bottom()
    cv2.rectangle(RGB_im,(x1,y1),(x2,y2),(220,255,220),1)
    if y1 < 0: y1 = 0 
    if y2 < 0: y2 = 0
    if x1 < 0: x1 = 0
    if x2 < 0: x2 = 0
    imgCrop = RGB_im[y1:y2, x1: x2]
    # plt.imshow(imgCrop)
    img = cv2.resize(imgCrop,(150,150))
    x = image.img_to_array(img)
    x = preprocess_input(x)
    x = np.expand_dims(x, axis=0)
    images = np.vstack([x])
    classes = model.predict(images, batch_size=32)
    # print(classes)
    if classes >= 0.5:
        print(file + ": " + 'mask\n')
        mask += 1
    else:
        print(file + ": " + 'not_mask\n')
        not_mask += 1
print("Total mask :",mask)
print("Total not_mask :",not_mask)
```
